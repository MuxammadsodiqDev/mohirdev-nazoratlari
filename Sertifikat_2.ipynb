# Datasetni modelga tayyorlash

```python
from sklearn.model_selection import StratifiedShuffleSplit
stratified_split = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in stratified_split.split(df,df['satisfaction']):
  train_set = df.iloc[train_index]
  test_set =df.loc[test_index]

X = train_set.drop('satisfaction',axis=1)
y = train_set['satisfaction']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder

num_pipeline = Pipeline([
    ('std_scaler',StandardScaler()),
    ('imputer',SimpleImputer(strategy='median'))
])
```

```python
from sklearn.compose import ColumnTransformer
num_attribs = ['Departure/Arrival time convenient','Arrival Delay in Minutes','Gate location','Departure Delay in Minutes','Age','Flight Distance','Inflight wifi service','Ease of Online booking','Food and drink','Online boarding','Seat comfort','Inflight entertainment' ,'On-board service','Leg room service'  ,'Baggage handling' ,'Checkin service' ,'Inflight service' ,'Cleanliness' ]
cat_attribs = ['Customer Type','Type of Travel','Class']


full_pipeline = ColumnTransformer([
    ('num',num_pipeline,num_attribs),
    ('cat',OneHotEncoder(),cat_attribs),
])

X_prepared =full_pipeline.fit_transform(X)
```

# Model (RandomForestClassifier)


```python
from sklearn.ensemble import RandomForestClassifier

RFC_model = RandomForestClassifier(n_estimators=500,min_samples_split=3)
RFC_model.fit(X_prepared,y)
```

# Modelni baholaymiz

```python
X_test = test_set.drop('satisfaction',axis=1)
y_test = test_set['satisfaction']
```

```python
X_test_prepared = full_pipeline.transform(X_test)
y_test_predict = RFC_model.predict(X_test_prepared)
```

```python
from sklearn import metrics
print(f"Model aniqligi: ",metrics.accuracy_score(y_test,y_test_predict))
```
